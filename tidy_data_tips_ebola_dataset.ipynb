```python
##Tidying data 
#Reference from Daniel Chen Pydata(2018)

'''
Let's see how we can make the dataset less 'wide' but 'long'?
We see that there are many repeats of similar records.
Can we somehow change this to record date,day,country, deaths clearly?
'''
```

```python
import pandas as pd

#Example 2 
ebolas_data = pd.read_csv('country_timeseries.csv')

ebola_long = pd.melt(ebolas_data,id_vars=['Date','Day'])
ebola_long.head()
```

```python
'''
Holding Date and Day results in something we would want. 
But we realise that the countries count,death lumped together 
'''

'''
How can we split case count, death counts?
Let's see what we can do with the data we have.
Lets try do something with 'Cases_Guinea'. Essentially this is a string 
Maybe we can try to split this string
'''
```

```python
#Select the column like calling a list
#Use the str accessor to do 'string' properties on the data
variable_col_split = ebola_long['variable'].str.split('_')

#Results in a Series type aka ' Column '
#Think in Python. A Series is like a python List. 
#Hence, we can access a list data like how we call through a list.
#In this example, our 'list' is like an array. or a list of list.
variable_col_split
```

```python
#To further iterate what i mean 
variable_col_split[0]
```

```python
#To further iterate what i mean #2
variable_col_split[0][0]
```

```python
#To further iterate what i mean #3
variable_col_split[0][1]
```

```python
#Or this 

a_list = [['abc','efg'],['hik',4],[5,6]]
for u in a_list:
    print(u[0])
```

```python
#But we can actually use the properties of the data which are strings to our advantage
variable_col_split.str[0]
```

```python
#Similarly, to access the countries we can do this 
variable_col_split.str[1]
```

```python
#Now that we have found out how to split up the data, all we need to do is to insert it into our dataframe
#If we want to create a new column for dataframe, simply treat it as though it is existent
ebola_long['Type'] = variable_col_split.str[0]
ebola_long['Country'] = variable_col_split.str[1]
ebola_long.head()
```

```python
#Alternate way to do all these in a single line 

#Making use of the 'expand' parameter of split method 
#We see that we get a dataframe as a result which defualt '0' , '1' for column names
ebola_long['variable'].str.split('_',expand = True)
```

```python
#Thus, we can make use of dataframe properties 

ebola_long[['Mirror_Type','Mirror_Country']] = ebola_long['variable'].str.split('_',expand = True)
ebola_long
```
